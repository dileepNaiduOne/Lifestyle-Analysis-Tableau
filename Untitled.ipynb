```python
import pandas as pd
```

```python
df = pd.read_csv("lifestyle.csv")
```

```python
df.head()
```

```python
females = df[df["Gender"] == "female"].sample(16000, random_state=523647)
```

```python
males = df[df["Gender"] == "male"].sample(4000, random_state=124)
```

```python
data = pd.concat([females, males], axis=0)
```

```python
data
```

```python
data["Gender"].value_counts()
```

```python
s = data[data["Country"] == "Turkey"].sample(402, random_state=7)
```

```python
data = data[~(data["Country"] == "Turkey")]
```

```python
data = pd.concat([data, s], axis=0)
```

```python
data["Country"].value_counts()
```

```python
data["Gender"].value_counts()
```

```python
data.to_csv("Data.csv", index=False)
```
